```python
import numpy as np
import matplotlib.pyplot as plt


np.random.seed(42)
n_samples = 100

X = np.random.rand(n_samples, 1) * 10
true_slope = 2.5
true_intercept = 5.0
noise = np.random.randn(n_samples, 1)
y = true_slope * X + true_intercept + noise

plt.scatter(X, y, color="blue", alpha=0.5, label="Data points")
plt.title("Synthetic Regression Dataset")
plt.xlabel("X")
plt.ylabel("y")
plt.legend()
plt.show()


w = np.random.randn(1)
b = np.random.randn(1)
print(f"Initial weight: {w}, Initial bias: {b}")


def predict(X, w, b):
   
    return X * w + b


def compute_loss(y_true, y_pred):
 
    return np.mean((y_true - y_pred) ** 2)


def gradient_descent(X, y, w, b, learning_rate):
   
    n = len(y)
    y_pred = predict(X, w, b)

    dw = -2 * np.sum(X * (y - y_pred)) / n
    db = -2 * np.sum(y - y_pred) / n

    w -= learning_rate * dw
    b -= learning_rate * db
    return w, b



learning_rate = 0.01
n_epochs = 1000
loss_history = []

for epoch in range(n_epochs):
    y_pred = predict(X, w, b)
    loss = compute_loss(y, y_pred)
    loss_history.append(loss)

    w, b = gradient_descent(X, y, w, b, learning_rate)

    if (epoch + 1) % 100 == 0:
        print(f"Epoch {epoch + 1}, Loss: {loss:.4f}")

# Plot training loss history
plt.plot(range(1, n_epochs + 1), loss_history, label="Loss")
plt.xlabel("Epoch")
plt.ylabel("MSE Loss")
plt.title("Training Loss over Epochs")
plt.legend()
plt.show()


X_test = np.array([[0], [2], [4], [6], [8], [10]])
y_test_pred = predict(X_test, w, b)

plt.scatter(X, y, color="blue", alpha=0.5, label="Training data")
plt.plot(X_test, y_test_pred, color="red", label="Model Predictions")
plt.title("Model Predictions on Test Data")
plt.xlabel("X")
plt.ylabel("y")
plt.legend()
plt.show()

print(f"Final weight: {w}, Final bias: {b}")
```
